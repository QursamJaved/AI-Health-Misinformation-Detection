```python
import pandas as pd
df_train = pd.read_csv('cleaned_train.csv')
print(df_train.head())
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

fake_text = ' '.join(df_train[df_train['label']==0]['clean_tweet'].dropna().astype(str))
wc = WordCloud(width=800, height=400, background_color='black').generate(fake_text)
plt.figure(figsize=(10,5))
plt.imshow(wc)
plt.axis('off')
plt.title('Misinformation Word Cloud')
plt.savefig('wordcloud_misinfo.png')
plt.show()
```
