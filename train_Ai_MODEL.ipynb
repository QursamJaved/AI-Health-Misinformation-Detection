 STEP 1: Import the tools we need

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix
import joblib
```

STEP 2: Load the dataset

```python
df_train = pd.read_csv('cleaned_train.csv')
df_val = pd.read_csv('cleaned_val.csv')
df_test = pd.read_csv('cleaned_test.csv')

print("Train shape:", df_train.shape)
print("Val shape:", df_val.shape)
print("Test shape:", df_test.shape)
```

 STEP 3: Check and Fix missing text values

```python
print(df_train['clean_tweet'].isnull().sum())
print(df_val['clean_tweet'].isnull().sum())
print(df_test['clean_tweet'].isnull().sum())

df_train = df_train.dropna(subset=['clean_tweet'])
df_val = df_val.dropna(subset=['clean_tweet'])
df_test = df_test.dropna(subset=['clean_tweet'])

df_train['clean_tweet'] = df_train['clean_tweet'].astype(str)
df_val['clean_tweet'] = df_val['clean_tweet'].astype(str)
df_test['clean_tweet'] = df_test['clean_tweet'].astype(str)
```

STEP 4: Convert text into numbers (TF-IDF)

```python
vectorizer = TfidfVectorizer(max_features=5000)

X_train = vectorizer.fit_transform(df_train['clean_tweet'])
X_val = vectorizer.transform(df_val['clean_tweet'])
X_test = vectorizer.transform(df_test['clean_tweet'])

y_train = df_train['label']
y_val = df_val['label']
y_test = df_test['label']
```

STEP 5: Train the SVM model

```python
model_svm = LinearSVC()
model_svm.fit(X_train, y_train)

val_preds_svm = model_svm.predict(X_val)
print("SVM — Validation Results:")
print(classification_report(y_val, val_preds_svm))
```

STEP 6: Test on the test set (final, honest check)

```python
final_preds = model_svm.predict(X_test)

print("Final Test Results (SVM):")
print(classification_report(y_test, final_preds))
print(confusion_matrix(y_test, final_preds))
```

```python
import joblib
joblib.dump(model_svm, 'health_misinfo_model.pkl')
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')

from google.colab import files
files.download('health_misinfo_model.pkl')
files.download('tfidf_vectorizer.pkl')
```
